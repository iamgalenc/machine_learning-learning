```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
col_names  = ['age', 'suka dogs', 'suka astronomi', 'astronot']
df = pd.read_csv('astronot.csv', names = col_names)
df
```

```python
feature = ['age', 'suka dogs', 'suka astronomi']
X = df[feature]
Y = df['astronot']
```

```python
clf = DecisionTreeClassifier()
clf.fit(X, Y)
```

```python
plt.figure(figsize=(30,20), facecolor='m')
fg = tree.plot_tree(clf, feature_names= feature, fontsize=30)
```

```python
print(clf.predict([[10, 0, 1]]))
```
